Linear regression by using Deep Neural network: Implement Boston housing price prediction
problem by Linear regression using Deep Neural network. Use Boston House price prediction
dataset.

```python
import time
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
data = pd.read_csv("1_boston_housing.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
# Split features and target variable
x = data.drop('MEDV',axis=1)
y = data['MEDV']
```

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
# Standardize the features
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
#model

# Build the model
model = Sequential()

model.add(Dense(128, input_shape=(13, ), activation='relu', name='dense_1'))
model.add(Dense(64, activation='relu', name='dense_2'))
model.add(Dense(1, activation='linear', name='dense_output'))

# Compile the model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
# Start time
start_time = time.time()

history = model.fit(x_train, y_train, epochs=50, batch_size=32, validation_split=0.2,verbose=1)

# End time
end_time = time.time()

# Calculate total training time
training_time = end_time - start_time
print("Total training time: {:.2f} seconds".format(training_time))
```

```python
# Predictions on test data
predictions = model.predict(x_test)

# Print the first few predictions
for i in range(5):
    print("Predicted Price:", predictions[i][0], "\tActual Price:", y_test.iloc[i])

# You can also calculate metrics such as mean squared error and mean absolute error on the test data
from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)

print('Mean Squared Error on test data: ', mse)
print('Mean Absolute Error on test data: ', mae)
```
